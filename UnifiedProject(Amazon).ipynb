```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/Amazon Sales data.csv")

# Display initial data insights
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())

# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract month and year from 'Order Date'
df['Month'] = df['Order Date'].dt.month
df['Year'] = df['Order Date'].dt.year

# Group data by month and sum the 'Units Sold'
monthly_sale = df.groupby('Month')['Units Sold'].sum().reset_index()

# 1. Display month-wise units sold
print("\nMonth-wise Units Sold:\n", monthly_sale)

plt.figure(figsize=(5, 5))
sns.barplot(x='Month', y='Units Sold', data=monthly_sale, palette='viridis')
plt.title('Month-wise Sales')
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.xticks(range(0, 12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()

# Group data by year and sum the 'Units Sold'
yearly_sale = df.groupby('Year')['Units Sold'].sum().reset_index()

# 2. Display year-wise units sold
print("\nYear-wise Units Sold:\n", yearly_sale)

plt.figure(figsize=(6, 6))
plt.pie(yearly_sale['Units Sold'], labels=yearly_sale['Year'], autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Yearly Sales')
plt.show()

# 3. Yearly month-wise sale
yearly_monthly_sale = df.groupby(['Year', 'Month'])['Units Sold'].sum().reset_index()
print("\nYearly Month-wise Sale:\n", yearly_monthly_sale)

# 4. Bar plot of items sold
item_sold = df.groupby('Item Type')['Units Sold'].sum().reset_index()

plt.figure(figsize=(5, 5))
sns.barplot(x='Item Type', y='Units Sold', data=item_sold, palette='coolwarm')
plt.xlabel('Item Type')
plt.ylabel('Total Units Sold')
plt.title('Bar Plot of Items Sold')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Sales according to channel
sales_channel = df.groupby('Sales Channel')['Units Sold'].sum().reset_index()

print("\nSales Channel-wise Sale:\n", sales_channel)

plt.figure(figsize=(5, 5))
sns.barplot(x='Sales Channel', y='Units Sold', data=sales_channel, palette='viridis')
plt.xlabel("Sales Channel")
plt.ylabel("Number of Units Sold")
plt.title('Sales by Channel')
plt.grid(True)
plt.show()

# Priority wise selling
priority_selling = df.groupby('Order Priority')['Order ID'].count().reset_index()
priority_selling.columns = ['Order Priority', 'Order Count']

print("\nPriority-wise Orders:\n", priority_selling)

plt.figure(figsize=(6, 6))
plt.pie(priority_selling['Order Count'], labels=priority_selling['Order Priority'], autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Order Priority Distribution')
plt.show()

# Product-wise profit
product_wise_profit = df.groupby('Item Type')['Total Profit'].sum().reset_index()

print("\nProduct-wise Profit:\n", product_wise_profit)

plt.figure(figsize=(5, 5))
sns.barplot(x='Item Type', y='Total Profit', data=product_wise_profit, palette='cubehelix')
plt.xlabel("Item Type")
plt.ylabel("Total Profit")
plt.title('Product-wise Profit')
plt.xticks(rotation=80)
plt.grid(True)
plt.tight_layout()
plt.show()

# Region-wise units sold
region_wise_sale = df.groupby('Region')['Units Sold'].sum().reset_index()
print("\nRegion-wise Units Sold:\n", region_wise_sale)

# Product-wise total cost
product_wise_total_cost = df.groupby('Item Type')['Total Cost'].sum().reset_index()

print("\nProduct-wise Total Cost:\n", product_wise_total_cost)

plt.figure(figsize=(5, 5))
sns.barplot(y='Item Type', x='Total Cost', data=product_wise_total_cost, palette='magma')
plt.title('Product-wise Total Cost')
plt.xlabel('Total Cost')
plt.ylabel('Item Type')
plt.show()
```
